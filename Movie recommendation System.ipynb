```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=movies.merge(credits,on='title')
movies.shape
```

```python
# useful columns : genre, id, keywords, title, overview, cast, crew
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(10)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated()
```

```python
movies.iloc[0].genres
```

```python
import ast
from ast import literal_eval
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj): #string will convert into list using ast.liternal_eval(obj)
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
import ast
from ast import literal_eval
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj): #string will convert into list using ast.liternal_eval(obj)
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
import ast
from ast import literal_eval
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj): #string will convert into list using ast.liternal_eval(obj)
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

# Vectorization : convert tag of each movie into 2d vector
text vectorization : In Vectorization we have to convert text into 2d vector and after that the system will recommend nearest vector's movie
we're going to use  bag of words vectorization technique in this project
    what is bag of words vectorization? : step1 : combine all tags -> it will become very large tag -> in this large scale of combination, system will calculate most common words, in this case i'm extracting most 5000 common words. after that system will check same for each row and create vector for each row. and then recommend most nearest movies. in this case i'm not considerin g stop words(is, am, are etc.)

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

# What is Stemming?
Stemming is the process of reducing words to their root or base form by removing suffixes. It helps in normalizing words so that variations of a word (e.g., "running", "runs", "ran") are reduced to a common base form (e.g., "run").

Example of Stemming
Word : 
Played	play,
Happily	happi,
Studies	studi.

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem('loving')
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
# calculates first movie's similarity with all other movies, 
# added index using enumerate(which is tuple (index,similarity)=(0,0.12)),
# after that converted tuple into list
# the most outsider method, sorted will sort the similarity according to index but we have to sort it according to similarity.
# for that i applied key=lambda x:x[1] that can sort the similarity not the indexes
# sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(i)
#         print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
